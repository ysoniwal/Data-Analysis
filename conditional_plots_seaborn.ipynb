# Conditional plots using seaborn for titanic disaster train dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

titanic = pd.read_csv("datasets/train_titanic.csv")
titanic = titanic[["Survived", "Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"]]
titanic = titanic.dropna()
titanic.head(1)
```

#### Set seaborn style, remains same in all seaborn plots  in current session style categories are
+ darkgrid
+ whitegrid
+ dark
+ white
+ tick

```python
sns.set_style("white")
sns.distplot(titanic["Age"])
plt.show()
```

```python
# KDE plots (Kernel Density Estimation)
sns.kdeplot(titanic["Age"], shade = True)

# Seaborn.despine(), without argument removes 
# right and top spines
sns.despine(left = True, bottom = True)

plt.xlabel("Age")
plt.show()
```

#### Conditional plots with one condition

```python
# Create a grid for each unique value of "Pclass"
g = sns.FacetGrid(titanic, col = "Pclass", size = 6)

# plot KDE plot for "Age" for each unique value of "Pclass"
g.map(sns.kdeplot, "Age", shade = True)

sns.despine(left = True, bottom = True)

plt.show()
```

#### Conditional plots with 2 conditions

```python
g = sns.FacetGrid(titanic, row = "Survived", col = "Pclass")
g.map(sns.kdeplot, "Age", shade = True)

sns.despine(left = True, bottom = True)
plt.show()
```

#### Conditional plots with 3 conditions

```python
g = sns.FacetGrid(titanic, row = "Survived", col = "Pclass", hue = "Sex", hue_order = ["female", "male"])
g.map(sns.kdeplot, "Age", shade = True).add_legend()

sns.despine(left = True, bottom = True)
plt.show()
```
